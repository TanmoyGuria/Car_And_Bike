```python
base_url = "https://www.carandbike.com/new-cars/models/{}"

car_names = []
images = []



for page_num in range(1, 12):
    url = base_url.format(page_num)
    soup = get_soup(url)
    
    if soup:
        div_tags = soup.find_all('div', class_='relative')
        for div in div_tags:
            img_tag = div.find('img')
            if img_tag and 'src' in img_tag.attrs:
                image_link = img_tag['src']
                images.append(image_link)
```

```python
import requests
from bs4 import BeautifulSoup

def get_soup(url):
    try:
        r = requests.get(url)
        if r.status_code == 200:
            return BeautifulSoup(r.content, 'html.parser')
    except Exception as e:
        print(f"Error fetching {url}: {e}")
    return None

url = "https://www.cardekho.com/maruti-suzuki-cars"
soup = get_soup(url)

image_links = []

if soup:
    img_tags = soup.find_all('img', class_='hover')
    for img in img_tags:
        if 'src' in img.attrs:
            image_link = img['src']
            image_links.append(image_link)

print(image_links)
```

```python
import requests
from bs4 import BeautifulSoup

# The URL of the page you want to scrape
url = "https://www.cartrade.com/maruti-suzuki-cars/"

# Send a request to fetch the HTML content
response = requests.get(url)
html_content = response.content

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')

# Find all <a> tags with the specific class
a_tags = soup.find_all('a', class_='car-list-model-image position-relative')

# Create lists to store image URLs and titles
image_urls = []
image_titles = []

# Loop through all the found <a> tags
for a in a_tags:
    # Find the <img> tag within the <a> tag
    img = a.find('img')
    
    # Extract the src (URL) and title from the <img> tag
    if img and img.get('src') and img.get('title'):
        image_url = img['src']
        image_title = img['title']
        
        # Append the extracted URL and title to the lists
        image_urls.append(image_url)
        image_titles.append(image_title)

df1 = pd.DataFrame({
    'Title': image_titles,
    'Image URL': image_urls
})

# Display the DataFrame
print(df1)

```
