```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
def get_soup(url):
    response = requests.get(url)
    return BeautifulSoup(response.content, 'html.parser')
```

```python
base_url = "https://www.carandbike.com/used/cars-in-new-delhi/{}"

all_car_names = []
km_driven = []
fuel_types = []
transmission_types = []
price=[]


for page_num in range(1, 8):
    url = base_url.format(page_num)
    soup = get_soup(url)
    
    if soup:
        name_tags = soup.find_all("div", class_="ucTitle truncate truncate1")
        for tag in name_tags:
            car_name = tag.text.strip()
            all_car_names.append(car_name)
        
        car_info_tags = soup.find_all("ul", class_="ucInfo flex pt10 pb10")
        for info in car_info_tags:
            items = info.find_all("li")
            if len(items) >= 3:
                km_driven.append(items[0].text.strip())
                fuel_types.append(items[1].text.strip())
                transmission_types.append(items[2].text.strip())
                
        price_tags = soup.find_all("div", class_="ucPrice flex gridJcgridAc")
        for tag in price_tags:
            car_price= tag.text.strip()
            price.append(car_price)


print(all_car_names)
print(km_driven)
print(fuel_types)
print(transmission_types)
print(price)
```

```python
data1 = {
    'Name': all_car_names,
    'Kilometers Driven': km_driven,
    'Fuel Type': fuel_types,
    'Transmission Type': transmission_types,
    'Price': price
}
```

```python
df1 = pd.DataFrame(data1)
df1['City'] = 'New Delhi'

df1
```
